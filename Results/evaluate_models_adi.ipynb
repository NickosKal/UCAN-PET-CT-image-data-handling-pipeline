## Age - Regression 

```python
cv_mae_arr = np.load("/home/ashish/Ashish/UCAN/Results/regression/Experiment_3/CV_0"+"/MAE.npy")
for i in range(1,10):
    temp = np.load("/home/ashish/Ashish/UCAN/Results/regression/Experiment_3/CV_"+str(i)+"/MAE.npy")
    cv_mae_arr = np.column_stack((cv_mae_arr,temp))
```

```python
cv_mae_df = pd.DataFrame(cv_mae_arr)
```

```python
fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(20,10))
fig.text(0.5, 0.04, 'Epochs', fontdict=dict(fontsize= 20), ha='center')
fig.text(0.04, 0.5, 'MAE', fontdict=dict(fontsize= 20), va='center', rotation='vertical')
colors = plt.rcParams["axes.prop_cycle"]()
cv=0
for i in range(2):
    for j in range(5):
        c = next(colors)["color"]
        axs[i, j].title.set_text(f'Fold: {cv}')
        axs[i, j].plot(range(cv_mae_arr.shape[0]), cv_mae_df.loc[:,cv].values, color = c)
        cv += 1

plt.show()
```
