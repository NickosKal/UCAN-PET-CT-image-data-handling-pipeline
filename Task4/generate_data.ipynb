```python
import pandas as pd
import numpy as np

def display_full(x):
    with pd.option_context("display.max_rows", None,
                           "display.max_columns", None,
                           "display.width", 20000,
                           "display.max_colwidth", None,
                           ):
        print(x)
```

```python
collages_with_nan_arrays = pd.read_excel("/media/andres/T7 Shield1/UCAN_project/df_of_arrays_with_nan_arrays.xlsx")
all_collages = pd.read_excel("/media/andres/T7 Shield1/UCAN_project/df_of_reshaped_collages.xlsx")

collages_with_nan_arrays = collages_with_nan_arrays.sort_values(by="scan_date")
collages_with_nan_arrays['scan_date'] = collages_with_nan_arrays['scan_date'].astype(str)
collages_with_nan_arrays['unique_patient_ID_scan_date'] = collages_with_nan_arrays['patient_ID'] + '_' + collages_with_nan_arrays['scan_date']
collages_with_nan_arrays = collages_with_nan_arrays.drop(columns=['patient_ID', 'scan_date', 'incorrect_projection'])

all_collages = all_collages.sort_values(by="scan_date")
all_collages['scan_date'] = all_collages['scan_date'].astype(str)
all_collages['unique_patient_ID_scan_date'] = all_collages['patient_ID'] + '_' + all_collages['scan_date']
all_collages = all_collages.drop(columns=['patient_ID', 'scan_date'])
```
